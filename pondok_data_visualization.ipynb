````python
# make a new dataframe where it contains all row of data except the rows with all NaN in its Untuk Dihubungi, Tel, E-mel, Tel & Faks, Laman web, Faks
pondok_data_detailed_filtered = pondok_data_detailed.dropna(subset=['Untuk dihubungi', 'Tel', 'E-mel', 'Tel & Faks', 'Laman web', 'Faks'], how='all')
pondok_data_detailed_filtered
````
